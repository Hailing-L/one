```python
import pandas as pd
import matplotlib.pyplot as plt

# 步骤 1：读取数据
df = pd.read_csv("mtcars.csv")

# 步骤 3：拆分为手动和自动挡汽车
manual_cars = df[df['am'] == 1]
automatic_cars = df[df['am'] == 0]

# 步骤 4：创建 1x3 图表
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Understanding the Effect of Horsepower on Other Features")

# 步骤 5 和 6：hp vs mpg
axes[0].scatter(manual_cars['hp'], manual_cars['mpg'], color='blue', label='Manual')
axes[0].scatter(automatic_cars['hp'], automatic_cars['mpg'], color='red', label='Automatic')
axes[0].set_title("HP vs MPG")
axes[0].set_xlabel("Horsepower (HP)")
axes[0].set_ylabel("Miles per Gallon (MPG)")

# 添加水平线（进阶任务）
axes[0].axhline(manual_cars['mpg'].mean(), color='blue', linestyle='--', label='Manual Avg MPG')
axes[0].axhline(automatic_cars['mpg'].mean(), color='red', linestyle='--', label='Automatic Avg MPG')

# 步骤 7：hp vs qsec
axes[1].scatter(manual_cars['hp'], manual_cars['qsec'], color='blue', label='Manual')
axes[1].scatter(automatic_cars['hp'], automatic_cars['qsec'], color='red', label='Automatic')
axes[1].set_title("HP vs QSEC")
axes[1].set_xlabel("Horsepower (HP)")
axes[1].set_ylabel("Quarter Mile Time (QSEC)")

# 步骤 8：hp vs disp
axes[2].scatter(manual_cars['hp'], manual_cars['disp'], color='blue', label='Manual')
axes[2].scatter(automatic_cars['hp'], automatic_cars['disp'], color='red', label='Automatic')
axes[2].set_title("HP vs DISP")
axes[2].set_xlabel("Horsepower (HP)")
axes[2].set_ylabel("Displacement (DISP)")

# 添加图例并保存图表
for ax in axes:
    ax.legend()
plt.savefig("horsepower_analysis.png")
plt.show()
```
